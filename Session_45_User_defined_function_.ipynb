# **Types of Functions based on Parameters in Python**

1. Function without Parameters

2. Function with Parameters

3. Function with default Parameters

```python
# Funtion  without parameters`
def greet():
  name = input("Enter your name :")
  print(f"Hello {name} welcome to Python Class")
  print("Good Evening")

greet()
```

```python
# Funtion  with parameters`
def greet(name):
  print(f"Hello {name} welcome to Python Class")
  print("Good Evening")

greet("Ravi")
# greet()
```

```python
# Funtion  with default parameters`
def greet(name = "Student"):
  print(f"Hello {name} welcome to Python Class")
  print("Good Evening")

greet("Ravi")

greet()
```

**Instructions**

Create a function called arth_cal()

Find addition

Find subtraction

Find multiplication

Return all three values using multiple return statements


Take two numbers from the user using input()

Store returned values in separate variables

```python
# Arithmetic Calculation using User Input

def arth_cal(a, b):

    s = a + b
    d = a - b
    m = a * b

    return s, d, m


# User Input
num1 = int(input("Enter First Number : "))
num2 = int(input("Enter Second Number : "))

# Function Call
sum, difference, multiplication = arth_cal(num1, num2)

# Output
print("\nSum of 2 numbers :", sum)
print("Difference of 2 numbers :", difference)
print("Multiplication of 2 numbers :", multiplication)
```

# **Function with Loops in Python**

```python
# sum of n numbers passed
def sum_n(n):
  total = 0
  for i in range(1,n+1):
    total  = total + i
  return total

sum_n(20)
```

# **Case Study – Student Marks Analysis using Function and Loop**

**Instructions**

Create a list called marks

Create a function called marks_analysis() and pass the marks list to the function

Inside the function:

Use a for loop to display all marks

Find total marks

Find average marks


Return total and average marks, call the function, and display: Student Marks, Total Marks,  Average Marks

```python
# Marks List
marks = [80, 75, 90, 85, 70]

# Function Creation
def marks_analysis(data):

    print("Student Marks :")

    # for loop
    for i in data:
        print(i)

    # Operations
    total = sum(data)
    average = total / len(data)

    # Return Values
    return total, average


# Function Call
total_marks, average_marks = marks_analysis(marks)

# Output
print("\nTotal Marks   :", total_marks)
print("Average Marks :", average_marks)
```

```python
def count(numbers):
  even_count = odd_count = 0

  for n in numbers:
    if n % 2 == 0:
      even_count += 1
    else:
      odd_count += 1

  return even_count , odd_count


seq_list = [10,20,34,53,56,57,89,65,78,35]

even , odd = count(seq_list)
print(f"Total numbers of odd number: {odd}  \ntotal numbers of even num:{even}")
```
