Case Study – Student Registration System
Instructions:

Create variables:

student_name

student_age

course_name

course_fee

Use input() function to get student details from the user.

Use type casting:

int() for student age

float() for course fee

Calculate:

gst = 18% of course fee

total_fee = course fee + gst

Display all student details using:

print()

f-string

Display datatype of important variables using:

type()

```python
#Case Study – Student Registration System Instructions:
# INPUT FORM STUDENTS:
student_name = input("Enter student name:")
student_age = int(input("Enter student age:"))
course_name = input("Enter course name:")
course_fee = float(input('Enter course fees:'))

# CALCULATE GST AND TOTAL_FEES:
gst = course_fee * 18/100
total_fee=course_fee + gst

# DISPLAY ALL STUDENT DETAILS:
print("\n-- Student Registration Details --")
print(f"Student Name:{student_name}")
print(f"Student Age:{student_age}")
print(f"Course Name:{course_name}")
print(f"Course Fees:{course_fee}")
print(f"GST:{gst}")
print(f"Total Fees:{total_fee}")

# DISPLAY DATA TYPES:

print("\n-- Data Types --")
print(f"Student Name:{type(student_name)}")
print(f"Student Age:{type(student_age)}")
print(f"Course Name:{type(course_name)}")
print(f"Course Fees:{type(course_fee)}")
print(f"GST:{type(gst)}")
print(f"Total Fees:{type(total_fee)}")
```

Case Study – Library Fine Calculator
Instructions:

Create variables:

student_name

book_name

days_late

fine_per_day

Use input() function to get details from the user.

Use type casting:

int() for days_late

float() for fine_per_day

Calculate:

total_fine = days_late * fine_per_day

Display all details using:

print()

f-string

Display datatype of important variables using:

type()

```python
# Case Study – Library Fine Calculator Instructions:

# INPUT FROM USERS:

student_name = input("Enter Student Name:")
book_name = input("Enter Book Name:")
days_late = int(input("Enter Days Late:"))
fine_per_day = float(input("Enter Fine Per day:"))

# CALCULATE TOTAL FINE:

total_fine = days_late * fine_per_day

# DISPLAY USERS DETAILS:

print("\n --Users Details--")
print(f"Student Name:{student_name}")
print(f"Book Name:{book_name}")
print(f"Days Late:{days_late}")
print(f"Fine Per Day:{fine_per_day}")
print(f"Total Fine:{total_fine}")

# DISPLAY DATA TYPES:

print(f"Student Name:{type(student_name)}")
print(f"Book Name:{type(book_name)}")
print(f"Days Late:{type(days_late)}")
print(f"Fine Per Day:{type(fine_per_day)}")
print(f"Total Fine:{type(total_fine)}")





```
