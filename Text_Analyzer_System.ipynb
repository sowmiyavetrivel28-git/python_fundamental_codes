Case Study – Text Analyzer System
Instructions:

Take a string input from the user using input().

Convert the text into lowercase using lower().

Use a for loop to iterate through each character in the string.

Count: • Total vowels
• Total consonants
• Total digits
• Total special characters

Use: • isalpha() for alphabets
• isdigit() for numbers

Display all counts using print()

```python
text=(input('enter your text:')).lower()

vowels=consonants=digits=specialchar=0

for letter in text:

  if letter in "aeiou":
    vowels+=1

  elif letter.isalpha():
    consonants+=1

  elif letter.isdigit():
    digits+=1
  else:
    specialchar+=1

print(f'total vowels in text             : {vowels}')
print(f"total consonent in text          : {consonants}")
print(f"total digits in text             : {digits}")
print(f"total special characters in text : {specialchar}")
```
