## **Case Study – Employee Attendance Count System using continue**

**Instructions:**

1. Create a list named `attendance` with values:
   Present, Absent, Present, Leave, Present

2. Create count variables:

   • present  
   • absent  
   • leave  

3. Use a for loop to iterate through the attendance list.

4. Use continue statement to:
   • Skip remaining conditions after counting Present employees  

5. Count:
   • Total Present employees  
   • Total Absent employees  
   • Total Leave employees  

6. Display all counts using print()

```python
attendance=["Present","Absent","Present","Leave","Present"]

Present=Absent=Leave=0


for status in attendance:

   if status=="Present":
     Present+=1
     continue

   if status=="Absent":
     Absent+=1

   if status=="Leave":
     Leave+=1

print(f"Total Present Employees : {Present}")
print(f"Total Absent Employees  : {Absent}")
print(f"Total Leave Employees   : {Leave}")

```
