# **Case Study – Student Marks Analysis using Function and Loop**

**Instructions**

Create a list called marks

Create a function called marks_analysis() and pass the marks list to the function

Inside the function:

Use a for loop to display all marks

Find total marks

Find average marks


Return total and average marks, call the function, and display: Student Marks, Total Marks,  Average Marks

```python
marks=[60,56,65,70,85,90]

def mark_analysis(marks):

    print(f"student marks: ")


    for i in marks:
      print(i)

    total=sum(marks)
    average=total/len(marks)

    return total,average

total_mark,avg_mark=mark_analysis(marks)
print(f"total marks of the students is: {total_mark}")
print(f'average marks of the students is: {avg_mark}')

```
