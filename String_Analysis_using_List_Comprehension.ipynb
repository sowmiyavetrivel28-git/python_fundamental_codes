Case Study – String Analysis using List Comprehension
**Instructions **

Create a string variable called text

Store any sentence or word inside the variable

Use list comprehension to:

Convert all characters into uppercase

Extract only vowels from the string

Extract only consonants from the string

Display:

Original string

Uppercase characters

Vowels list

Consonants list

String without spaces

Print the output in a clean format using print() statements

```python
text="sowmiyavetrivel"

upper_text=[x.upper() for x in text]

vowels=[x for x in text if x in "aeiou"]

consonents=[x for x in text if x not in "aeiou"]

remove_space=[x for x in text if x!=" "]

print(f"original text : {text}")
print(f"upper text    : {upper_text}")
print(f"vowels        : {vowels}")
print(f"consonents    : {consonents}")
print(f'without_space : {remove_space}')
```
