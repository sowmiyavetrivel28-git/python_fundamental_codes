**Strings (Concatenation, Slicing, and Other Methods) :**

**1. String Concatenation:**

```python
string1 = "Hello"
string2 = input("Enter Your Name:")
concate_string = string1 + " " +string2
print(f"{concate_string}")
string3 ="Welcome to Python programming"
final_message = string1 +" "+string2+","+string3
print(f"{final_message}")
```

**2. String Slicing and Indexing:**



```python
# a. Print the first character of the string.

final_message[0]
print(f"The first character of string:{final_message[0]}")
```

```python
# b. Print the last character of the string.

final_message[40]
print(f"The last character of the string:{final_message[40]}")
```

```python
# c. Print the first 5 characters of the string.

final_message[0:5]
print(f"The first 5 characters of the string:{final_message[0:5]}")
```

```python
#  d. Print the last 11 characters of the string.

final_message[29:]
print(f"The last 11 characters of the string:{final_message[29:]}")
```

```python
# e. Print the string in reverse.

final_message[::-1]
print(f"Print the string in reverse:{final_message[::-1]}")
```

```python
#  f. Use slicing and print the word “Python” from the existing string.

final_message[23:29]
print(f"Use slicing and print the word “Python” from the existing string:{final_message[23:29]}")
```

**3. String Methods:**

```python
# Write a Python program that takes a string, strM = “Python beginner tutorial

strM = "Python beginner tutorial"
print(f"{strM}")
```

```python
# a. Convert the sentence to uppercase.

strM.upper()
print(f"{strM.upper()}")
```

```python
# b. Convert the sentence to lowercase.

strM.lower()
print(f"{strM.lower()}")
```

```python
# c. Use Capitalize and return the sentence to the original input form.

strM.capitalize()
print(f"{strM.capitalize()}")
```

```python
# d. Count the total number of occurrences of character ‘t’ in the string.

strM.count("t")
print(f"the total number of occurrences of character ‘t’:{strM.count("t")}")
```

```python
# e. Replace all occurrences of “Python” with “Data Analytics” in the input string

strM.replace("Python","Data Analytics")

print(f"Replace all occurrences of “Python” with “Data Analytics”:{strM.replace("Python","Data Analytics")}")
```

**4.Tuples (Creation, Modification and Access) :**

```python
# Create the 1st tuple with values -> (10, 20, 30) and the 2nd tuple with values -> (40, 50, 60):

tuple1=(10,20,30)
tuple2=(40,50,60)
```

```python
# a. Concatenate the two tuples and store it in “t_combine”

t_combine=tuple1+tuple2
print(f"t_combine:{t_combine}")
```

```python
# b. Repeat the elements of “t_combine” 3 times

repeat=t_combine*3
print(f" Repeat the elements of “t_combine” 3 times:{repeat}")
```

```python
# c. Access the 3rd element from “t_combine”

t_combine[2]
print(f"the 3rd element from “t_combine”:{t_combine[2]}")
```

```python
# d. Access the first three elements from “t_combine”

t_combine[0:3]
print(f" the first three elements from t_combine:{t_combine[0:3]}")
```

```python
# e. Access the last three elements from “t_combine”

t_combine[3:]   # or [3:6]
print(f"the last three elements from “t_combine:{t_combine[3:]}")
```
