Case Study – Arithmetic Calculator using User Defined Function
Instructions

Create a function called arth_cal()

Pass two parameters to the function

Inside the function:Perform addition, subtraction, multiplication

Return all three values using multiple return statements

Take two numbers from the user using input()

Store returned values in separate variables

Display:Sum, Difference, Multiplication

```python
def arthmetic_calc (a,b):

     sum=a+b
     sub=a-b
     mul=a*b
     div=a/b
     return sum,sub,mul,div

num1=int(input("Enter your num1: "))
num2=int(input('Enter your num2: '))

suma,suba,mula,diva = arthmetic_calc(num1,num2)

print(f"sum of numbers: {suma}")
print(f'sub of numbers: {suba}')
print(f"mul of numbers: {mula}")
print(f"div of numbers: {diva}")
```
