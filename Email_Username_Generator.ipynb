Case Study – Email Username Generator
Instructions:

• Create variables: full_name, domain_name

• Take input from the user using input() function

• Convert full_name into lowercase using lower()

• Remove spaces in name and replace with dot using replace()

• Split full_name using split() function

• Join name parts using join() function if needed

• Create email in format: name@domain

• Display only Full Name and Generated Email

• Use print() and f-string for output display

```python


full_name = input("enter your name:")
domain_name = input("enter your domain name:")

lower_name = full_name.lower()
email_name = lower_name.replace(' ','.')

email = email_name + domain_name +"@"+".com"

print("full_name:",full_name)

print('emailid:',email)

```

Case Study – Student ID Card Generator Instructions:

• Create variables: student_name, roll_number, department

• Take input from the user using input() function

• Extract first 4 letters of student_name and department using slicing

```python
# Case Study – Student ID Card Generator Instructions:
student_name = input("enter student name:")
roll_number = int(input("enter your roll number:"))
department = input("enter your department:")

upper_name = student_name.upper()
upper_dept = department.upper()
print(student_name[0:5])
print(department[0:4])

student_id=f"{upper_name}{roll_number}{upper_dept}"
print(f"student name:{upper_name}")
print(f'roll number:{roll_number}')
print(f'department:{upper_dept}')
print(f'student id:{student_id}')
```
