Case Study Scholarship Eligibility Checker
Instructions:

Create variables: student_name, marks, attendance, family_income.

Use conditional statements (if, elif, else) to:

Award Full Scholarship if marks ≥ 90 and attendance ≥ 90.
Award Half Scholarship if marks ≥ 80 or attendance ≥ 85.
Award Need-Based Scholarship if marks ≥ 70 and family_income < 30000.
Otherwise, display Not Eligible.
Display all data (no loops).

```python
student_name=input("Enter your name:")
marks=int(input("Enter your mark:"))
attendance=int(input("Enter your attendance:"))
family_income=int(input("Enter your family income:"))

if marks>=90 and attendance>=90:
   award= "full scholarship"

elif marks>=80 or attendance>=85:
   award="half scholarship"

elif marks>=70 and family_income<30000:
  award ="need_based scholarship"
else:
  print('not eligible')

print(f"student name:{student_name}")
print(f"mark:{marks}")
print(f"attendance:{attendance}")
print(f"family income:{family_income}")
print(f"award status:{award}")
```
