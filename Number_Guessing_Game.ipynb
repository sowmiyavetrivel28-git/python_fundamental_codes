```python
import random
secret_number=random.randint(1,10)


attempts=3

while attempts >0:
  guess_number=int(input("Enter your number:"))

  if  guess_number <1 or guess_number >10:
    print(f"Your guess is out of range. Please guess a number between 1 and 10.")
    continue

  if guess_number==secret_number:
    print(f"Congratulations! You guessed the correct number.")
    break

  elif guess_number < secret_number:
    print(f"Too low. Try again.")


  elif guess_number > secret_number:
    print(f"Too high. Try again.")


  attempts-=1
  print(f"attempts left: {attempts}")

else:
    print(f"Better luck next time!")

    print("The secret number was:", secret_number)

```
