# **Case Study – Student Performance Management System**

**Instructions**

Create two lists:
students
marks

Use range() to:
Access and print student names and marks using index positions

Use enumerate() to:
Display student index number and student name

Use zip() to:
Combine student names and marks together

Display all outputs using for loops

Use proper headings with print() statements

Print the final result in a clean format

```python
students=["livi","vidhun","Tamil"]
marks=[90,95,92]

print(f"\n---- using range() ----\n")

for i in range(len(students)):
  print(f"{students[i]} {"-"} {marks[i]}")

print(f"\n---- using enumerate() ----\n")

for index,name in enumerate(students):
  print(f"{index} {"-"} {name}")

print(f"\n---- using zip() ----\n")

for name,score in zip(students,marks):
  print(f"{name} {"-"} {score}")
```
