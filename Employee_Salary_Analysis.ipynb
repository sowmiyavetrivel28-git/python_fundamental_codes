Case Study – Employee Salary Analysis
Instructions

Create a list called salary Store employee salaries inside the list

Use Built-in Functions to:

Find maximum salary

Find minimum salary

Find total salary

Find number of employees using len()

Use lambda with map() to: Increase each salary by 10%

Use lambda with filter() to: Display salaries greater than 50000

Display:

Original salary list

Updated salary list

High salary list

Maximum salary

Minimum salary

Total salary

```python
salary=[30000,45000,55000,75000]

max_salary=max(salary)
min_salary=min(salary)
total_salary=sum(salary)
no_of_employees=len(salary)

salary_increment=list(map(lambda s : (s*0.10)+s,salary))

greater_salary=list(filter(lambda s:s>50000,salary))


print(f"original salary list       : {salary}")

print(f"updated salary list        : {salary_increment}")

print(f'high salary list           : {greater_salary}')

print(f"maximum salary in the list : {max_salary}")

print(f"minimum salary in the list : {min_salary}")

print(f"total salary in the list   : {total_salary}")

print(f"no of employees            : {no_of_employees}")
```
