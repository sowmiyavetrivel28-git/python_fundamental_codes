
1. Create a dictionary named `student` with the following keys:
   name, age, course

2. Store multiple students' data using lists as values.

3. Add one new student details to the dictionary:

   • Append new name to name list  

   • Append new age to age list  

   • Append new course to course list  

4. Display all student details using dictionary keys.

5. Access and display:

   • First student's details using indexing

   • Newly added student's details using indexing

```python
student= {
    "name":["sowmi","vetri","vidhun"],
    "age":[31,32,2],
    "course":["DA","Software","KG"]
}

# ADD NEW STUDENT
student["name"].append("sona")
student["age"].append("30")
student['course'].append("CA")

# DISPLAY ALL STUDENT DETAILS
print(f"All names:{student["name"]}")
print(f"All ages:{student["age"]}")
print(f"All course:{student["course"]}")

# first student details
print("\n --first student details--")
print(f"name:{student["name"][0]}")
print(f"age:{student['age'][0]}")
print(f'course:{student['course'][0]}')

# newly added student deatails

print("\n --new student details--")
print(f"name:{student["name"][3]}")
print(f"age:{student['age'][3]}")
print(f"course:{student['course'][3]}")
```

Case Study – Student Course Registration System using Set
Instructions:

Create a set named courses with course names: Python, SQL, Power BI, Python, Excel

Perform set operations: • Add course using add()
• Add multiple courses using update()
• Remove course using remove() and discard()

Display all courses and total number of courses using len()

```python
courses={"python","sql","powerBI","python","Excel"}

# ADD COURSES
courses.add("java")
print(f"affter add:{courses}")

#add multiple courses
courses.update(["javascript",'react'])
print(f"after update:{courses}")

# remove
courses.remove("java")
print(f"after remove:{courses}")

# discard

courses.discard("sql")
print(f'after discard:{courses}')

# display all courses
print(f"All courses:{courses}")

#len()
len(courses)

```
