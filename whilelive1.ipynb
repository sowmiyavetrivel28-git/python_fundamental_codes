Case Study – Password Verification System
Instructions:

Create a variable named password with a default password value.

Take password input from the user using input().

Use a while loop to: • Check whether the entered password matches the original password
• Display "Incorrect password... try again" if password is wrong
• Ask user to re-enter password until correct password is entered

Display "Access granted" when the correct password is entered

```python
# Case Study – Password Verification System

# Original password
password = "python123"

# Taking input from user
user_password = input("Enter the password: ")

# while loop for password checking
while user_password != password:
    print("Incorrect password... try again")
    user_password = input("Enter the password: ")

else:
    print("Access granted")
```

Case Study – Password Verification System with Limited Attempts
Instructions:

Create variables: • password
• attempts
• max_attempts

Use a while loop to allow the user to enter the password.

Check whether the entered password matches the original password.

If the password is correct: • Display "Access granted!"
• Stop the loop using break

If the password is incorrect: • Increase attempt count
• Display remaining attempts

If maximum attempts are reached: • Display "Maximum attempts reached. Access denied!"

```python
password="adminsowmi"
attempts=0
max_attempts=3

while attempts< max_attempts:
  user_password=input("Enter your password:")

  if user_password==password:
    print("Access granted!")
    break
  else:
    attempts+=1
    print(f"remaining attempts:{max_attempts-attempts}")

if attempts== max_attempts:
   print("maximum attempts reached.Access denied")
```

```python
password = "adminsowmi"
attempts = 0
max_attempts = 3

while attempts < max_attempts:
    user_password = input("Enter your password: ")

    if user_password == password:
        print("Access granted!")
        break
    else:
        attempts += 1
        print(f"Remaining attempts: {max_attempts - attempts}")

if attempts == max_attempts:
    print("Maximum attempts reached. Access denied.")
```

Case Study – Student Marks Filter System using continue
Instructions:

Create two lists: • students
• marks

Use a while loop to iterate through the student list.

Check student marks using if condition.

Use continue statement to: • Skip students who scored 50 or above

Display only students who scored below 50.

Print student name and marks using f-string








```python
students=['sowmi',"vetri",'vidhun','livi']
marks=[45,98,49,95]

i=0

while i < len(students):

    if marks[i] >= 50:
      i+=1
      continue

print(f'student  {students[i]} mark is {marks[i]}')
i+=1
```
