# **List**[]

•	A mutable ordered collection of items.


**Features**


1.	Ordered → Elements keep their order.


2.	Mutable → Can add, remove, or modify elements.


3.	Allows duplicates → Same value can appear multiple times.


4.	Heterogeneous → Can store integers, strings, floats, booleans, even other lists.

```python
# Initial list
fruits = ["Apple", "Banana", "Cherry", "Apple"]
print(type(fruits))
```

```python
# modify the elements
fruits[2]="orange"
print(f'frutis:{fruits}')
```

```python
# append() - Add item at end
fruits.append("Mango")
print("After append():", fruits)
```

```python
# append()
fruits.append('mango')
print(f"fruits:{fruits}")
```

```python
#insert(index,value) add elements in specific position
fruits.insert(3,"watermelon")
print(f"insert fruits:{fruits}")
```

```python
# extend() add multiple elements at end
fruits.extend(["kiwi","muskmelon"])
print(f"extend fruits:{fruits}")
```

```python
# remove() remove first accurance of value

print(f'old fruits:{fruits}')
fruits.remove("Apple")
print(f'after remove:{fruits}')
```

```python
#pop() remove last item
fruits.pop()
print(f"fruits:{fruits}")
```

```python
fruits.pop(0)
print(f"fruits pop with index:{fruits}")
```

```python
# count() count accurance of value
fruits.count("orange")
```

```python
fruits.index("watermelon")
print(f"fruits:{fruits.index('watermelon')}")
```

```python
#sort() sort the list
fruits.sort()
print(f"sort fruits:{fruits}")
```

```python
# reverse () reverse the list
fruits.reverse()
print(f"reverse fruits:{fruits}")
```

```python
#sort() sort the list
fruits.sort(reverse=True)
print(f"reverse using sort fruits:{fruits}")
```

# **Case Study – Shopping Cart System**

**Instructions:**

1. Create a list named `cart` with initial items:
   Milk, Bread, Eggs, Apple

2. Perform list operations:

   • Add multiple items using extend()

   • Insert item at index 2 using insert()

   • Remove item Apple using remove()

   • Remove last item using pop()

3. Use list methods:

   • index() to find position of Eggs

   • count() to count occurrences of Milk

4. Apply indexing:

   • Display first item

   • Display last item

5. Apply slicing:

   • Display first 3 items from the list

6. Find total number of items using len()

7. Display all outputs using print()

```python
cart=["Milk", "Bread","Egg","Apple"]

print(f"\nshopping list:{cart}")

cart.extend(["paneer","Mushroom"])
print(f"\nextend cart:{cart}")

cart.insert(2,"Gee")
print(f"\ninserted cart:{cart}")

cart.remove("Apple")
print(f"\n after remove:{cart}")

cart.pop()
print(f'\n after pop:{cart}')

cart.index("Egg")
print(f"\n index position of the egg is:{cart.index("Egg")}")

cart.count("Milk")
print(f"\n count of milk:{cart.count("Milk")}")

cart[0]
print(f"\n first item:{cart[0]}")

cart[4]
print(f"\n last item:{cart[4]}")


cart[:3]
print(f"\n first three items:{cart[:3]}")

len(cart)
print(f"\n total len of cart:{len(cart)}")


```
